```python
corr_data = df[['views_millions', 'likes_millions', 'revenue_millions', 'imdb_rating']]
corr_matrix = corr_data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Views, Likes, Revenue and Rating')
plt.show()
```

- This heatmap shows the correlation between views ,likes, revenue, and IMDB rating.
- all correlation values are very close to 0 that means there is no relationship between these variable
- views and likes have a very weak positive correlation
- views and revenue show a weak negative correlation meaning higher views do not always mean higher revenue
- revenue and IMDB rating have a weak positive correlation

- drama genre has high content count in English and Malayalam
- documentary content is highly available in Hindi and also good in other languages
- Sci-Fi genre is popular in Tamil and Telugu
- romance has the lowest count in Kannada show less romance content in that language
